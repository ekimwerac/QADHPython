Series
1. Create a series from the below list:
attempts = [1, 3, 2, 3, 2, 3, 1, 1, 2, 1]
2. Find the first person’s number of attempts.
3. Find the first five people’s number of attempts.
4. Find the last three people’s number of attempts.
Here's how you can accomplish each of the tasks using NumPy and Pandas:

```python
import pandas as pd

# Create a series from the list
attempts = [1, 3, 2, 3, 2, 3, 1, 1, 2, 1]
attempts_series = pd.Series(attempts)

# Find the first person’s number of attempts
first_attempt = attempts_series.iloc[0]
print("First person's number of attempts:", first_attempt)

# Find the first five people’s number of attempts
first_five_attempts = attempts_series.iloc[:5]
print("First five people's number of attempts:\n", first_five_attempts)

# Find the last three people’s number of attempts
last_three_attempts = attempts_series.iloc[-3:]
print("Last three people's number of attempts:\n", last_three_attempts)
```

Output:
```
First person's number of attempts: 1
First five people's number of attempts:
 0    1
1    3
2    2
3    3
4    2
dtype: int64
Last three people's number of attempts:
 7    1
8    2
9    1
dtype: int64
```

Explanation:
1. We create a Pandas Series from the list `attempts`.
2. To find the first person's number of attempts, we use `.iloc[0]`, which selects the first element of the series.
3. To find the first five people's number of attempts, we use `.iloc[:5]`, which selects the first five elements of the series.
4. To find the last three people's number of attempts, we use `.iloc[-3:]`, which selects the last three elements of the series.

````python
import pandas as pd

# Create a series from the list
attempts = [1, 3, 2, 3, 2, 3, 1, 1, 2, 1]
attempts_series = pd.Series(attempts)

# Find the first person’s number of attempts
first_attempt = attempts_series.iloc[0]
print("First person's number of attempts:", first_attempt)

# Find the first five people’s number of attempts
first_five_attempts = attempts_series.iloc[:5]
print("First five people's number of attempts:\n", first_five_attempts)

# Find the last three people’s number of attempts
last_three_attempts = attempts_series.iloc[-3:]
print("Last three people's number of attempts:\n", last_three_attempts)
````

# DataFrames
1. Create and display a DataFrame from the below dictionary and list of index labels.
```python
exam_data = {'name': ['Anne', 'Kofi', 'Ridhwaan', 'Zara', 'Muhammad', 'Donald', 'Isabella', 'Sarah', 'Paula', 'John'],
             'score': [12.5, 9, 16.5, None, 9, 20, 14.5, None, 8, 19],
             'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1], 
             'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
labels = ['Student ' + letter for letter in ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']]
```
2. Display the first three rows of the data frame.
3. Display the name and score columns of the data frame.
4. Display the name and score columns and rows 1, 3, 4, and 8.
5. Display the rows where the number of attempts is greater than 2.
6. Display the rows where score is greater or equal than a pass rate of 10.

````python
import pandas as pd

# Define the exam data dictionary and index labels
exam_data = {'name': ['Anne', 'Kofi', 'Ridhwaan', 'Zara', 'Muhammad', 'Donald', 'Isabella', 'Sarah', 'Paula', 'John'], 
             'score': [12.5, 9, 16.5, None, 9, 20, 14.5, None, 8, 19], 
             'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1], 
             'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}

labels = ['Student ' + letter for letter in ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']]

# Create DataFrame from the dictionary with index labels
df = pd.DataFrame(exam_data, index=labels)

# Display the DataFrame
print("DataFrame:")
print(df)

# Display the first three rows of the DataFrame
print("\nFirst three rows:")
print(df.head(3))

# Display the name and score columns of the DataFrame
print("\nName and score columns:")
print(df[['name', 'score']])

# Display the name and score columns and rows 1, 3, 4, and 8
print("\nName and score columns, rows 1, 3, 4, and 8:")
print(df.loc[['Student b', 'Student d', 'Student e', 'Student h'], ['name', 'score']])

# Display the rows where the number of attempts is greater than 2
print("\nRows where the number of attempts is greater than 2:")
print(df[df['attempts'] > 2])

# Display the rows where score is greater or equal than a pass rate of 10
print("\nRows where score is greater or equal than a pass rate of 10:")
print(df[df['score'] >= 10])
# Display the rows where the score is above 15 and the student has only made one attempt
print("\nRows where score is above 15 and the student has only made one attempt:")
print(df[(df['score'] > 15) & (df['attempts'] == 1)])
````

# Querying DataFrames -
1. Compute the monthly earnings of mortgage applicants, just using Income.
2. Compute the ratio of debts to assets for the mortgage applicants.
3. Display all mortgage applicants who have a Balance greater than £1000.
4. Display all mortgage applicants who have a Balance greater than £1000 and a Debt below £50.
5. Display all loan applicants who have an Income greater than 30,000 who have a 10-year loan, and those with an Income greater than 20,000 who have a 20-year loan (together!)


````python
import pandas as pd

# Load the Mortgage applicants data from file
df=pd.read_csv("data/mortgage_applicants.csv")

# 1. Compute the monthly earnings of mortgage applicants, just using Income.
monthly_earnings = df['Income'] / 12
print("Monthly earnings of mortgage applicants:")
print(monthly_earnings)

# 2. Compute the ratio of debts to assets for the mortgage applicants.
debt_to_assets_ratio = df['Debt'] / df['Balance']
print("\nRatio of debts to assets for mortgage applicants:")
print(debt_to_assets_ratio)

# 3. Display all mortgage applicants who have a Balance greater than £1000.
balance_gt_1000 = df[df['Balance'] > 1000]
print("\nMortgage applicants with Balance greater than £1000:")
print(balance_gt_1000)

# 4. Display all mortgage applicants who have a Balance greater than £1000 and a Debt below £50.
balance_gt_1000_debt_lt_50 = df[(df['Balance'] > 1000) & (df['Debt'] < 50)]
print("\nMortgage applicants with Balance greater than £1000 and Debt below £50:")
print(balance_gt_1000_debt_lt_50)

# 5. Display all loan applicants who have an Income greater than 30,000 who have a 10-year loan, 
# and those with an Income greater than 20,000 who have a 20-year loan (together!)
loan_applicants = df[((df['Income'] > 30000) & (df['Term'] == '10 Years')) |
                     ((df['Income'] > 20000) & (df['Term'] == '20 Years'))]
print("\nLoan applicants with specified criteria:")
print(loan_applicants)
````

# Aggregating DataFrames
1. Compute the average Balance of mortgage applicants depending on the Term of their loan.
2. Compute the average Income of mortgage applicants depending on whether they defaulted or not.
3. Compute the average Debt, Income, and Balance of mortgage applicants based on whether they defaulted or not, and the term of their loan.
4. Add a column to the DataFrame called MeanTermIncome, which contains the mean Income of mortgage applicants based on their Term.

````python
import pandas as pd

# Load the Mortgage applicants data from file
df=pd.read_csv("data/mortgage_applicants.csv")

# 1. Compute the average Balance of mortgage applicants depending on the Term of their loan.
average_balance_by_term = df.groupby('Term')['Balance'].mean()
print("Average Balance of mortgage applicants depending on the Term of their loan:")
print(average_balance_by_term)

# 2. Compute the average Income of mortgage applicants depending on whether they defaulted or not.
average_income_by_default = df.groupby('Default')['Income'].mean()
print("\nAverage Income of mortgage applicants depending on whether they defaulted or not:")
print(average_income_by_default)

# 3. Compute the average Debt, Income, and Balance of mortgage applicants based on whether they defaulted or not, 
# and the term of their loan.
average_stats_by_default_and_term = df.groupby(['Default', 'Term']).agg({'Debt': 'mean', 'Income': 'mean', 'Balance': 'mean'})
print("\nAverage Debt, Income, and Balance of mortgage applicants based on whether they defaulted or not, and the term of their loan:")
print(average_stats_by_default_and_term)

# 4. Add a column to the DataFrame called MeanTermIncome, which contains the mean Income of mortgage applicants based on their Term.
df['MeanTermIncome'] = df.groupby('Term')['Income'].transform('mean')
print("\nDataFrame with MeanTermIncome column:")
print(df)
````
