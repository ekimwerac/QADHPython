# Exercise 8 - Methods for Visualising Data

Load in the dataset `train_viz.csv`

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('data/train_viz.csv')
```

Create a scatter plot of `duration` vs. `price`. Segment by `destination`.

```python
plt.scatter(x=df['duration'],
            y=df['price'])
plt.xlabel('duration (m)')
plt.ylabel('price (Â£)')
```

```python
sns.scatterplot(data=df,
               x='duration',
               y='price',
               hue='destination')
```

Create a barplot of `fare` counts

```python
plt.bar(x=df['fare'].unique(), 
       height=df['fare'].value_counts())
plt.ylabel('price (Â£)')
plt.xticks(rotation=60);
```

```python
sns.countplot(data=df,
              x='fare')
plt.xticks(rotation=60);
```

Create a box & whisker plot of `price` with respect to `destination`

```python
plt.boxplot(x=df['price']);
```

```python
sns.boxplot(data=df,
           x='destination',
           y='price')
```

Create a histogram of `duration`

```python
plt.hist(df['duration']);
```

```python
sns.histplot(data=df,
            x='duration')
```

Compare this to a histogram of `price`

```python
sns.histplot(data=df,
            x='price')
```

Create a violin plot of price with respect to destination

```python
sns.violinplot(data=df,
           x='destination',
           y='price')
```

Create a heatmap of the correlation between columns

```python
sns.heatmap(df.corr(numeric_only=True),
            cmap='coolwarm',
            annot=True,
            center=0.0)
```

Create a proportional stacked bar plot of `vehicle_class` by `destination`

```python
fig, ax = plt.subplots(figsize=(8, 5))
table = pd.crosstab(df['destination'], df['vehicle_class'])
table.div(table.sum(1).astype(float), axis=0).plot(kind='bar', stacked=True, ax=ax);
```
