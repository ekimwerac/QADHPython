### Aggregating DataFrames

Compute the average `Balance` of mortgage applicants depending on the `Term` of their loan

```python
mortgage[['Balance', 'Term']].groupby('Term').mean()
```

Compute the average `Income` of mortgage applicants depending on whether they defaulted or not

```python
mortgage[['Income', 'Default']].groupby('Default').mean()
```

Compute the average `Debt`, `Income`, and `Balance` of mortgage applicants based on whether they defaulted or not, and the term of their loan

```python
mortgage[['Debt', 'Income', 'Balance', 'Default']].groupby('Default').mean()
```

Add a column to the DataFrame called MeanTermIncome, which contains the mean `Income` of mortgage applicants based on their `Term`. 

```python
mortgage['MeanTermIncome'] = mortgage[['Income', 'Term']].groupby('Term').transform(np.mean).round(2)
mortgage
```
