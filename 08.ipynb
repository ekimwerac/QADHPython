Creating a basic graph in matplotlib involves several steps, from importing the necessary libraries to plotting the data. Here's a detailed explanation of how to do it, including the use of `plt.subplots()`.

### Step-by-Step Guide

1. **Import Necessary Libraries**:
   - Import `matplotlib.pyplot` for plotting.
   - Import `numpy` for creating sample data (optional, but useful for generating data points).

2. **Create Sample Data**:
   - Use `numpy` to generate some data points for plotting.

3. **Create a Plot**:
   - Use `plt.subplots()` to create a figure and axis for the plot.
   - Plot the data on the axis.

4. **Customize the Plot**:
   - Add titles, labels, and legends to make the plot informative.

5. **Display the Plot**:
   - Use `plt.show()` to display the plot.

### Example Code

Here's a complete example that demonstrates these steps:

```python
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Create sample data
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.sin(x)                # sin(x) values for the points

# Step 2: Create a plot
fig, ax = plt.subplots()

# Step 3: Plot the data
ax.plot(x, y, label='sin(x)')

# Step 4: Customize the plot
ax.set_title('Sine Wave')          # Add a title
ax.set_xlabel('x values')          # Add an x-axis label
ax.set_ylabel('sin(x) values')     # Add a y-axis label
ax.legend()                        # Add a legend

# Step 5: Display the plot
plt.show()
```

### Explanation of the Code

1. **Imports**:
   - `import matplotlib.pyplot as plt`: This imports the `pyplot` module from matplotlib and aliases it as `plt`.
   - `import numpy as np`: This imports numpy and aliases it as `np`, useful for generating numerical data.

2. **Create Sample Data**:
   - `x = np.linspace(0, 10, 100)`: Generates 100 evenly spaced points between 0 and 10.
   - `y = np.sin(x)`: Calculates the sine of each point in `x`.

3. **Create a Plot**:
   - `fig, ax = plt.subplots()`: Creates a figure (`fig`) and a single set of axes (`ax`). `plt.subplots()` is useful because it can create multiple subplots in a single figure, but here we are just creating one.

4. **Plot the Data**:
   - `ax.plot(x, y, label='sin(x)')`: Plots `x` vs. `y` on the `ax` axes object and adds a label for the legend.

5. **Customize the Plot**:
   - `ax.set_title('Sine Wave')`: Sets the title of the plot.
   - `ax.set_xlabel('x values')`: Sets the label for the x-axis.
   - `ax.set_ylabel('sin(x) values')`: Sets the label for the y-axis.
   - `ax.legend()`: Adds a legend to the plot using the label provided in `ax.plot()`.

6. **Display the Plot**:
   - `plt.show()`: Displays the plot in a window.

### Summary

Using `matplotlib` and `numpy`, you can create and customize basic plots easily. The `plt.subplots()` function is particularly useful for creating a figure and one or more subplots, providing flexibility for more complex plotting needs.

````python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Step 1: Create sample data
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.sin(x)                # sin(x) values for the points

# Step 2: Create a plot
fig, ax = plt.subplots(figsize=(5,1))

# Step 3: Plot the data
ax.plot(x, y, label='sin(x)')

# Step 4: Customize the plot
ax.set_title('Sine Wave')          # Add a title
ax.set_xlabel('x values')          # Add an x-axis label
ax.set_ylabel('sin(x) values')     # Add a y-axis label
ax.legend()                        # Add a legend

# Step 5: Display the plot
plt.show()

#fig, ax = plt.subplots(figsize=(10,2))
````

## Plotting Data on Axes

Editing in OO Style


````python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df=pd.read_csv("data/loan_data.csv")
fig, (ax1,ax2) = plt.subplots(1,2, figsize=(10,2))
ax1.scatter(df[df['Term'] == 'Long Term']['Debt'],
            df[df['Term'] == 'Long Term']['Balance'])
ax1.set_xlabel('Debt ()')
ax1.set_ylabel('Balance ()')
ax1.set_title('Long Term')
ax1.set_xlim(-200, 14_000)
ax1.set_ylim(-200, 6_200)

ax2.scatter(df[df['Term'] == 'Short Term']['Debt'],
            df[df['Term'] == 'Short Term']['Balance'])
ax2.set_xlabel('Debt ()')
ax2.set_ylabel('Balance ()')
ax2.set_title('Short Term')
ax2.set_xlim(-200, 14_000)
ax2.set_ylim(-200, 6_200)
````

## Seaborn - makes plotting Easier
- it is still matplotlib under the hood


````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv("data/loan_data.csv")

# Create the FacetGrid plot
g = sns.FacetGrid(data=df, col='Term')
g.map(sns.scatterplot, 'Debt', 'Balance')

# Display the plot
plt.show()
````
